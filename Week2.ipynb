```python

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk import bigrams, trigrams
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from collections import Counter

print("Libraries loaded ")
```

```python
# Court sentences with anonymized names
sentences = [
    "Party A filed a suit against Party B for breach of contract.",
    "The court found Party B guilty of unlawful termination of services.",
    "The judge ruled that Party A failed to provide sufficient evidence.",
    "The High Court dismissed the case with costs to the respondent.",
    "Party B denied all allegations of wrongful suspension.",
    "Counsel for Party A submitted that the account was unjustly frozen.",
    "The court considered the pleadings and submissions of both parties.",
]

full_text = " ".join(sentences)
tokens = word_tokenize(full_text.lower())

print(f"Total tokens: {len(tokens)}")
print("Sample tokens:", tokens[:10])
```

```python
# Bigram

bigram_list  = list(bigrams(tokens))
bigram_freq  = Counter(bigram_list)

print("BIGRAM OUTPUT")
print("=" * 50)
print(f"{'Bigram':<40} {'Count'}")
print("-" * 50)
for bg, freq in bigram_freq.most_common(10):
    print(f"{str(bg):<40} {freq}")
```

```python
#POS Tagging Output

sample_sentence = "The court ruled that Party A failed to provide sufficient evidence."
tokens_pos      = word_tokenize(sample_sentence)
pos_tags        = pos_tag(tokens_pos)

# Simple tag explanations for students
tag_meanings = {
    'DT' : 'Determiner (the, a)',
    'NN' : 'Common Noun',
    'NNP': 'Proper Noun',
    'NNS': 'Plural Noun',
    'VBD': 'Verb - Past Tense',
    'VBZ': 'Verb - Present',
    'VB' : 'Verb - Base Form',
    'IN' : 'Preposition',
    'JJ' : 'Adjective',
    'TO' : 'Particle - to',
    'CC' : 'Conjunction',
    'PRP': 'Pronoun',
    'RB' : 'Adverb',
}

print("POS TAGGING OUTPUT")
print(f"Sentence: {sample_sentence}")
print("=" * 60)
print(f"{'Word':<20} {'POS Tag':<12} {'What it means'}")
print("-" * 60)
for word, tag in pos_tags:
    meaning = tag_meanings.get(tag, tag)
    print(f"{word:<20} {tag:<12} {meaning}")
```

```python
# Language Prediction using Trigrams

from nltk import trigrams
from nltk.tokenize import word_tokenize
from collections import Counter

# Rebuild the text and trigrams in this same cell
sentences = [
    "Party A filed a suit against Party B for breach of contract.",
    "The court found Party B guilty of unlawful termination of services.",
    "The judge ruled that Party A failed to provide sufficient evidence.",
    "The High Court dismissed the case with costs to the respondent.",
    "Party B denied all allegations of wrongful suspension.",
    "Counsel for Party A submitted that the account was unjustly frozen.",
    "The court considered the pleadings and submissions of both parties.",
]

full_text    = " ".join(sentences)
tokens       = word_tokenize(full_text.lower())
trigram_freq = Counter(trigrams(tokens))

# Prediction function
def predict_next_word(prefix, trigram_freq):
    prefix = tuple(prefix)
    candidates = {k: v for k, v in trigram_freq.items() if k[:2] == prefix}
    if not candidates:
        return "No prediction found"
    best = max(candidates, key=candidates.get)
    return best[2]

# Test predictions
test_prefixes = [
    ["the", "court"],
    ["party", "a"],
    ["the", "judge"],
    ["all", "allegations"],
]

print("LANGUAGE PREDICTION OUTPUT")
print("=" * 50)
print(f"{'Input Prefix':<30} {'Predicted Next Word'}")
print("-" * 50)
for prefix in test_prefixes:
    prediction = predict_next_word(prefix, trigram_freq)
    print(f"{str(prefix):<30} →  '{prediction}'")
```

```python
# Sentence Analysis Output

print("SENTENCE STRUCTURE ANALYSIS")
print("=" * 55)

for i, sentence in enumerate(sentences, 1):
    tokens_s    = word_tokenize(sentence)
    tags_s      = pos_tag(tokens_s)
    nouns       = [w for w, t in tags_s if t in ('NN','NNS','NNP','NNPS')]
    verbs       = [w for w, t in tags_s if t.startswith('VB')]
    adjectives  = [w for w, t in tags_s if t == 'JJ']

    print(f"\nSentence {i}: {sentence}")
    print(f"  Nouns      : {nouns}")
    print(f"  Verbs      : {verbs}")
    print(f"  Adjectives : {adjectives}")
```
