```python
# EXTRA WORK (JurisAI Case Classifier)

print("=" * 55)
print("   JURISAI — COURT CASE CLASSIFIER ")
print("=" * 55)

# Step 1: Define keyword rules for each case type
case_rules = {
    "Criminal"     : ["guilty", "convicted", "sentence", "fraud", "theft", "murder", "arrest", "bail"],
    "Civil"        : ["plaintiff", "defendant", "damages", "contract", "breach", "compensation", "sued"],
    "Land"         : ["land", "property", "title deed", "eviction", "ownership", "acres", "plot"],
    "Constitutional": ["petition", "rights", "constitution", "electoral", "commission", "freedom"],
    "Commercial"   : ["company", "bank", "appeal", "business", "shares", "trade", "corporation"],
}

# Step 2: Court cases to classify
court_cases = [
    {"id": "CASE-001", "text": "The plaintiff sued the defendant for breach of contract and claimed damages."},
    {"id": "CASE-002", "text": "The accused was convicted of fraud and sentenced to 5 years in prison."},
    {"id": "CASE-003", "text": "The petition challenged the electoral commission over violation of constitutional rights."},
    {"id": "CASE-004", "text": "The dispute involved ownership of land and eviction from the property."},
    {"id": "CASE-005", "text": "The bank appealed against the ruling over a commercial trade dispute."},
]

# Step 3: Classification function
def classify_case(text, rules):
    text_lower = text.lower()
    scores = {}
    for case_type, keywords in rules.items():
        score = sum(1 for keyword in keywords if keyword in text_lower)
        scores[case_type] = score
    best_match = max(scores, key=scores.get)
    return best_match, scores[best_match]

# Step 4: Classify all cases
print()
print(f"{'Case ID':<12} {'Predicted Type':<18} {'Confidence':<12} {'Text Preview'}")
print("-" * 75)

results = []
for case in court_cases:
    predicted_type, confidence = classify_case(case["text"], case_rules)
    results.append({
        "id"  : case["id"],
        "type": predicted_type,
        "conf": confidence,
        "text": case["text"]
    })
    print(f"{case['id']:<12} {predicted_type:<18} {confidence} keyword(s)    {case['text'][:35]}...")

# Step 5: Summary
print()
print("CLASSIFICATION SUMMARY")
print("=" * 40)
from collections import Counter
type_counts = Counter(r["type"] for r in results)
for case_type, count in type_counts.items():
    print(f"  {case_type:<18} : {count} case(s)")

print()
print("=" * 55)
print("Extra Work Complete ")
print("Technique Used : Keyword-Based Text Classification")
print("Application    : Automated Court Case Categorization")
print("Project        : JurisAI — Kenyan Court Document Analysis")
print("=" * 55)
```
