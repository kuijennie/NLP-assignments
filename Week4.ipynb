```python
# Dependency Parsing — Court Sentence

sentence = "The plaintiff filed a case against the defendant in court."
doc = nlp(sentence)

print("DEPENDENCY PARSING OUTPUT")
print("=" * 50)
print(f"Sentence: {sentence}")
print()
print(f"{'Word':<20} {'Role':<20} {'Depends On'}")
print("-" * 50)
for token in doc:
    print(f"{token.text:<20} {token.dep_:<20} {token.head.text}")
```

```python
# Subject, Verb and Object

doc = nlp("The judge dismissed the case.")

print("SENTENCE STRUCTURE OUTPUT")
print("=" * 40)
for token in doc:
    if token.dep_ == "nsubj":
        print(f"Subject (Who)  : {token.text}")
    if token.dep_ == "ROOT":
        print(f"Verb   (Action): {token.text}")
    if token.dep_ == "dobj":
        print(f"Object (What)  : {token.text}")
```

```python
# Semantic Similarity (Unrelated Sentences)

sent3 = nlp("The judge delivered the final judgment.")
sent4 = nlp("The weather in Nairobi is warm and sunny.")

score2 = sent3.similarity(sent4)

print("SEMANTIC SIMILARITY — UNRELATED SENTENCES")
print("=" * 50)
print(f"Sentence 1 : {sent3}")
print(f"Sentence 2 : {sent4}")
print(f"Score      : {score2:.4f}")
print(f"Meaning    : {'Similar' if score2 > 0.7 else 'Not Similar '}")
print()
print("OBSERVATIONS:")
print("-" * 50)
print("1. The score is LOW because the two sentences")
print("   have completely different topics.")
print("2. Sentence 1 is about a legal court ruling.")
print("3. Sentence 2 is about weather in Nairobi.")
print("4. They share no common legal vocabulary.")
print("5. A low score means the sentences are unrelated.")
print()
print("CONCLUSION: Legal sentences score higher because")
print("they share the same topic and vocabulary.")
print("Unrelated sentences score lower because they")
print("have nothing in common in meaning or context.")
```

```python
# Semantic Similarity — Similar Sentences

sent1 = nlp("The plaintiff filed a case against the defendant.")
sent2 = nlp("The claimant submitted a complaint against the respondent.")

score = sent1.similarity(sent2)

print("SEMANTIC SIMILARITY — SIMILAR SENTENCES")
print("=" * 50)
print(f"Sentence 1 : {sent1}")
print(f"Sentence 2 : {sent2}")
print(f"Score      : {score:.4f}")
print(f"Meaning    : {'Similar ' if score > 0.7 else 'Not Similar'}")
print()
print("OBSERVATIONS:")
print("-" * 50)
print("1. The score is HIGH because both sentences")
print("   describe the same legal action.")
print("2. 'Plaintiff' and 'claimant' mean the same thing.")
print("3. 'Filed a case' and 'submitted a complaint'")
print("   express the same action differently.")
print("4. Both sentences share legal vocabulary and context.")
print("5. A high score means the sentences are closely related.")
print()
print("CONCLUSION: Even though different words are used,")
print("the meaning is the same — making the similarity")
print("score high. This is how semantic analysis works.")
```
