```python
# Statistical Translation Table — Legal Terms

translation_table = {
    "court"      : ("mahakama",    0.97),
    "plaintiff"  : ("mdai",        0.95),
    "defendant"  : ("mshtakiwa",   0.93),
    "judge"      : ("jaji",        0.98),
    "case"       : ("kesi",        0.96),
    "evidence"   : ("ushahidi",    0.94),
    "ruling"     : ("uamuzi",      0.91),
    "dismissed"  : ("imekataliwa", 0.89),
    "guilty"     : ("na hatia",    0.92),
    "judgment"   : ("hukumu",      0.95),
    "lawyer"     : ("wakili",      0.97),
    "witness"    : ("shahidi",     0.93),
}

print("STATISTICAL TRANSLATION TABLE — Legal Terms")
print("=" * 50)
print(f"{'English':<15} {'Swahili':<18} {'Probability'}")
print("-" * 50)
for english, (swahili, prob) in translation_table.items():
    print(f"{english:<15} {swahili:<18} {prob:.2f}")
```

```python
# Simple Statistical Translator for Court Sentences

def simple_translate(sentence, table):
    words      = sentence.lower().split()
    translated = []
    for word in words:
        clean = word.strip(".,;:")
        if clean in table:
            swahili_word, prob = table[clean]
            translated.append(swahili_word)
        else:
            translated.append(word)
    return " ".join(translated)

test_sentences = [
    "The plaintiff filed a case against the defendant.",
    "The judge dismissed the case after reviewing the evidence.",
    "The lawyer submitted an appeal against the ruling.",
]

print("SIMPLE TRANSLATION OUTPUT — Court Sentences")
print("=" * 55)
for sentence in test_sentences:
    translation = simple_translate(sentence, translation_table)
    print(f"\nEnglish : {sentence}")
    print(f"Swahili : {translation}")
```
