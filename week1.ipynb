```python
# Tokenization Output

from nltk.tokenize import word_tokenize, sent_tokenize

# Sample court document
court_text = """
IN THE HIGH COURT AT CITY A
Civil Case No. 101 of 2023

BETWEEN:
PARTY A  ................  PLAINTIFF
AND
PARTY B  ................  DEFENDANT

JUDGMENT

The Plaintiff filed a case against the Defendant for breach of contract.
The Defendant denied all allegations and filed a defence.
The court considered all evidence and submissions from both parties.
Judgment is entered in favour of the Defendant.
The Plaintiff shall bear the costs of this suit.
"""

# Sentence Tokenization
sentences = sent_tokenize(court_text)
print("SENTENCE TOKENIZATION OUTPUT")
print("=" * 40)
print(f"Total sentences found: {len(sentences)}")
print()
for i, sentence in enumerate(sentences, 1):
    print(f"Sentence {i}: {sentence.strip()}")

print()

# Word Tokenization
words = word_tokenize(court_text)
print("WORD TOKENIZATION OUTPUT")
print("=" * 40)
print(f"Total words found: {len(words)}")
print()
print("First 20 words:", words[:20])
```

```python
# Stopwords Removal Output

from nltk.corpus import stopwords
import string

stop_words = set(stopwords.words('english'))

filtered_words = [
    word.lower() for word in words
    if word.lower() not in stop_words
    and word not in string.punctuation
    and word.isalpha()
]

print("STOPWORDS REMOVAL OUTPUT")
print("=" * 40)
print(f"Original word count   : {len(words)}")
print(f"After stopwords removed: {len(filtered_words)}")
print(f"Words removed         : {len(words) - len(filtered_words)}")
print()
print("Remaining meaningful words:")
print(filtered_words)
```
