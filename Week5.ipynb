```python
import nltk
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag

print("Week 5 ready ")
```

```python
# Complete NLP Pipeline

text = "The plaintiff filed a case against the defendant in the High Court."

print("COMPLETE NLP PIPELINE")
print("=" * 50)

# Step 1
tokens = word_tokenize(text)
print(f"Step 1 — Tokenization:")
print(f"  {tokens}")

# Step 2
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w.lower() not in stop_words and w.isalpha()]
print(f"\nStep 2 — Stopword Removal:")
print(f"  {filtered}")

# Step 3
pos = pos_tag(filtered)
print(f"\nStep 3 — POS Tagging:")
for word, tag in pos:
    print(f"  {word:<20} → {tag}")
```

```python
# Mini Legal Chatbot

print("=" * 45)
print("  JurisAI Legal Chatbot ")
print("=" * 45)

responses = {
    "hello"    : "Hello! How can I help you with your legal query?",
    "plaintiff": "A plaintiff is the person who files a case in court.",
    "defendant": "A defendant is the person being sued in court.",
    "judgment" : "A judgment is the final decision made by the court.",
    "appeal"   : "An appeal is a request to a higher court to review a decision.",
    "bye"      : "Goodbye! Visit JurisAI again for legal help.",
}

test_inputs = ["hello", "plaintiff", "judgment", "appeal", "bye"]

for user_input in test_inputs:
    print(f"\nYou : {user_input}")
    response = "I do not have information on that yet."
    for keyword, reply in responses.items():
        if keyword in user_input.lower():
            response = reply
            break
    print(f"Bot : {response}")
```
