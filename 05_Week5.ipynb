```python
import nltk
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag

print("Week 5 ready ")
```

```python
# Complete NLP Pipeline

text = "The plaintiff filed a case against the defendant in the High Court."

print("COMPLETE NLP PIPELINE")
print("=" * 50)

# Step 1
tokens = word_tokenize(text)
print(f"Step 1 — Tokenization:")
print(f"  {tokens}")

# Step 2
stop_words = set(stopwords.words('english'))
filtered = [w for w in tokens if w.lower() not in stop_words and w.isalpha()]
print(f"\nStep 2 — Stopword Removal:")
print(f"  {filtered}")

# Step 3
pos = pos_tag(filtered)
print(f"\nStep 3 — POS Tagging:")
for word, tag in pos:
    print(f"  {word:<20} → {tag}")
```

```python
# MINI PROJECT (CHATBOT)

import nltk
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

legal_responses = {
    "plaintiff"  : "A plaintiff is the party who initiates a court case.",
    "defendant"  : "A defendant is the party against whom a case is filed.",
    "judgment"   : "A judgment is the final decision delivered by the court.",
    "appeal"     : "An appeal is a request to a higher court to review a ruling.",
    "evidence"   : "Evidence is information presented in court to support a claim.",
    "bail"       : "Bail is temporary release of an accused person before trial.",
    "lawyer"     : "A lawyer is a professional who represents parties in court.",
    "case"       : "A case is a legal dispute brought before a court for resolution.",
    "court"      : "A court is a legal institution where disputes are resolved.",
    "ruling"     : "A ruling is a decision or order made by a judge.",
    "witness"    : "A witness is a person who gives evidence in a court case.",
    "sentence"   : "A sentence is the punishment given to a convicted person.",
    "acquitted"  : "Acquitted means the accused was found not guilty by the court.",
    "petition"   : "A petition is a formal legal request submitted to the court.",
    "injunction" : "An injunction is a court order stopping someone from doing something.",
}

case_rules = {
    "Criminal"      : ["guilty", "convicted", "sentence", "fraud", "theft", "murder", "arrest", "bail"],
    "Civil"         : ["plaintiff", "defendant", "damages", "contract", "breach", "compensation"],
    "Land"          : ["land", "property", "eviction", "ownership", "plot", "title"],
    "Constitutional": ["petition", "rights", "constitution", "electoral", "freedom"],
    "Commercial"    : ["company", "bank", "business", "shares", "trade", "corporation"],
}

def jurisai_respond(user_input):
    stop_words = set(stopwords.words('english'))
    tokens     = word_tokenize(user_input.lower())
    keywords   = [w for w in tokens if w not in stop_words and w.isalpha()]
    response   = None
    for keyword in keywords:
        if keyword in legal_responses:
            response = legal_responses[keyword]
            break
    scores    = {ct: sum(1 for k in kws if k in keywords) for ct, kws in case_rules.items()}
    case_type = max(scores, key=scores.get) if max(scores.values()) > 0 else "General"
    if not response:
        response = "I do not have information on that. Please consult a qualified lawyer."
    return response, case_type

test_inputs = [
    "what is a plaintiff",
    "what does judgment mean",
    "can someone apply for bail",
    "the defendant was found guilty of fraud",
    "what happens with land eviction cases",
]

print("=" * 55)
print("   JURISAI — LEGAL AID CHATBOT ")
print("=" * 55)

for user_input in test_inputs:
    response, case_type = jurisai_respond(user_input)
    print(f"\nYou      : {user_input}")
    print(f"JurisAI  : {response}")
    print(f"Case Type: {case_type}")
    print("-" * 55)
```

```python
# MINI PROJECT :INTERACTIVE CHATBOT

print("=" * 55)
print("   JURISAI — INTERACTIVE LEGAL CHATBOT ")
print("   Type a legal question. Type 'bye' to exit.")
print("=" * 55)
print()

while True:
    user_input = input("You: ")
    if user_input.lower() == "bye":
        print("JurisAI: Goodbye! Visit JurisAI again ")
        break
    if user_input.strip() == "":
        print("JurisAI: Please type a question.")
        continue
    response, case_type = jurisai_respond(user_input)
    print(f"JurisAI  : {response}")
    print(f"Case Type: {case_type}")
    print()
```
